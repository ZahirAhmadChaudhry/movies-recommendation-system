# Movies Recommendation System

Recommendation systems are machine learning techniques that help users discover relevant items or content based on their preferences, behavior, or similarity with other users. There are three main types of recommendation systems:

- Content-based: This type of system recommends items or content that are similar to the ones the user has liked, rated, or interacted with in the past. The system uses the features or attributes of the items or content to measure their similarity and rank them accordingly. For example, a content-based system can recommend movies to a user based on the genres, actors, directors, or ratings of the movies they have watched before.

- Collaborative filtering: This type of system recommends items or content that are popular or liked by other users who have similar tastes or preferences as the current user. The system uses the ratings, feedback, or actions of the users to learn their preferences and predict what they will like. For example, a collaborative filtering system can recommend books to a user based on the books that other users with similar reading habits have enjoyed.

- Hybrid: This type of system combines the strengths of both content-based and collaborative filtering systems to overcome their limitations and provide more accurate and diverse recommendations. The system can use different methods to integrate the two approaches, such as weighted voting, feature augmentation, or model blending. For example, a hybrid system can recommend music to a user based on both the features of the songs (such as genre, artist, or mood) and the ratings or playlists of other users who have similar music tastes.

### Pre-Processing on features

We have to perform followings on:

- From genres column, we only need genres names e.g. Action, Adventure etc.
- From Keywords, we only need keyword names e.g. culture clash, ocean, comics etc.
- From cast, we only need first three main charactors real names.
- From crew, we only need director name. 

We will perform pre-processing on each column individually becuase we have to extract the different information from each column. 

#### Genres

For genres, we need a function which iterate through it each dictionary and look for name, append it into list and return the list. 

```python
movies['genres'][0]
```

we can see that in genres column is a combination of dictionaries within a list, and it's represented as a string in JSON format. 
So, we need a function that converts a string representation of a list of dictionaries into a list of names. For example, if the text is `"[{'name': 'Alice', 'age': 20}, {'name': 'Bob', 'age': 25}]"`, the function will return `['Alice', 'Bob']`.

To do this, we need to use the `ast.literal_eval` function from the `ast` module. This function safely evaluates a string containing a Python literal or container display, such as lists, tuples, dicts, sets, etc. It returns the corresponding Python object that the string represents.

The reason why you need to use `ast.literal_eval` instead of the built-in `eval` function is that `eval` can execute any arbitrary Python code, which can be dangerous if the string comes from an untrusted source. `ast.literal_eval` only allows a limited set of Python structures, and raises an exception if the string contains anything else.

```python
def extract_genres(inp):
    list = []
    for i in ast.literal_eval(inp):
        list.append(i['name']) 
    return list
```

```python
movies['genres'] = movies['genres'].apply(extract_genres) #applying the function to the column genres
movies.head()
```

#### Keywords

Same as above.

```python
def extract_keywords(inp):
    list = []
    for i in ast.literal_eval(inp):
        list.append(i['name']) 
    return list
```

```python
movies['keywords'] = movies['keywords'].apply(extract_keywords) #applying the function to the column keywords
movies.head()
```

### Cast

As diccussed erlier, we only need first three characters from each movie.

```python
def extract_characters(inp):
    list = []
    counter = 0
    for i in ast.literal_eval(inp):
        if counter < 3:
            list.append(i['name'])
        counter+=1
    return list
```

```python
movies['cast'] = movies['cast'].apply(extract_characters) #applying the function to the column cast
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

#### Crew

From crew, we only interested in the director name. 

```python
def extract_director(inp):
    list = []
    for i in ast.literal_eval(inp):
        if i['job'] == 'Director':
            list.append(i['name'])
    return list
```

```python
movies['crew'] = movies['crew'].apply(extract_director) #applying the function to the column crew
movies.head()
```

We are done with desired information extration from desired columns. Now, we can see that in genres, keywords, cast, and crew we have words with spaces. When we will convert them into tags, each words will be treated as single tag e.g. for Sam Mendes tags will be #Sam #Mendes. That is a problem. So, we need to remove the spaces between words of these columns. Here how we will do that. 

```python
def no_spaces(inp):
    list_1 = []
    for i in inp:
        list_1.append(i.replace(" ",""))
    return list_1
```

```python
movies['cast'] = movies['cast'].apply(no_spaces)
movies['crew'] = movies['crew'].apply(no_spaces)
movies['genres'] = movies['genres'].apply(no_spaces)
movies['keywords'] = movies['keywords'].apply(no_spaces)
```
